```python
def parse_sector_column(series: pd.Series) -> pd.Series:
    """
    Try to turn a sector column into numeric seconds.

    Handles:
    - already-numeric columns
    - strings with ';' where one token is a time-like string ('1:24.567')
    - plain 'M:SS.mmm' strings
    """
    s = series.astype(str)

    # 1) Try numeric directly
    try:
        return pd.to_numeric(s)
    except Exception:
        pass

    # 2) If there are semicolons, split and pick the time-like bits
    if s.str.contains(";").any():
        tokens = s.str.split(";")

        def pick_time_token(row_tokens):
            for t in row_tokens:
                t = t.strip()
                if ":" in t:  # looks like time
                    return t
            return row_tokens[-1].strip()

        time_like = tokens.apply(pick_time_token)
    else:
        time_like = s

    # 3) Try parse as timedelta "M:SS.mmm"
    td = pd.to_timedelta(time_like, errors="coerce")
    if td.notna().any():
        return td.dt.total_seconds()

    # 4) Final fallback: numeric with NaN for bad values
    return pd.to_numeric(time_like, errors="coerce")
```

```python
def auto_pick_vir_columns(df: pd.DataFrame):
    cols = list(df.columns)

    # car number column candidates
    car_candidates = [
        c for c in cols
        if any(k in c.upper() for k in ["NUMBER", "CAR", "VEHICLE", "ENTRY"])
    ]

    # S1/S2/S3 candidates – any col name containing those tokens
    s1_candidates = [c for c in cols if "S1" in c.upper()]
    s2_candidates = [c for c in cols if "S2" in c.upper()]
    s3_candidates = [c for c in cols if "S3" in c.upper()]

    print("Car candidates:", car_candidates)
    print("S1 candidates:", s1_candidates)
    print("S2 candidates:", s2_candidates)
    print("S3 candidates:", s3_candidates)

    if not car_candidates or not s1_candidates or not s2_candidates or not s3_candidates:
        raise ValueError("Could not auto-detect car/S1/S2/S3 columns – check df.columns")

    car_col = car_candidates[0]
    s1_col = s1_candidates[0]
    s2_col = s2_candidates[0]
    s3_col = s3_candidates[0]

    print("Using columns:")
    print("  car_col =", repr(car_col))
    print("  s1_col  =", repr(s1_col))
    print("  s2_col  =", repr(s2_col))
    print("  s3_col  =", repr(s3_col))

    return car_col, s1_col, s2_col, s3_col
```

```python
def build_sector_stats_auto(df_sec: pd.DataFrame) -> pd.DataFrame:
    car_col, s1_col, s2_col, s3_col = auto_pick_vir_columns(df_sec)

    df = df_sec.copy()

    df["S1_s"] = parse_sector_column(df[s1_col])
    df["S2_s"] = parse_sector_column(df[s2_col])
    df["S3_s"] = parse_sector_column(df[s3_col])

    grouped = df.groupby(car_col, dropna=False)
    sector_stats = grouped[["S1_s", "S2_s", "S3_s"]].mean().reset_index()
    sector_stats = sector_stats.rename(columns={car_col: "car_number"})
    return sector_stats
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(1, 3, figsize=(15, 4), sharey=False)
sectors = ["S1_s", "S2_s", "S3_s"]
titles  = ["S1 mean time (s)", "S2 mean time (s)", "S3 mean time (s)"]

for ax, sec, title in zip(axes, sectors, titles):
    # boxplot of all cars
    ax.boxplot(sector_stats_r1[sec].dropna(), positions=[1], widths=[0.6])

    # our car's value
    our_val = sector_stats_r1.loc[sector_stats_r1["is_our_car"], sec].iloc[0]
    ax.scatter([1], [our_val], color="red", zorder=3, label="Car {}".format(CAR_NUM))

    ax.set_xticks([])
    ax.set_title(title)
    ax.set_ylabel("time (s)")

fig.suptitle(f"VIR – Race 1 sector mean times (Car {CAR_NUM} in red)")
plt.legend()
plt.tight_layout()
plt.show()
```

- **Context:** VIR – Race 1, sector mean times across the field, with **Car #2** shown as a red dot in each boxplot.

- **S1 (weakest):**
  - Field is tightly clustered around **48–49 s**.
  - Car #2’s dot is **slightly above the median**, meaning it’s a bit **slower than the typical competitive car** here.
  - This is the **main opportunity sector** for time gain (focus on braking, turn-in, and exit traction).

- **S2 (solid / slightly positive):**
  - Most cars sit around **52–53 s**.
  - Car #2 is **just below the median**, so it’s **slightly faster than the average**, but not the outright best.
  - Pace is **good but not dominant**; small refinements can help but S1 matters more.

- **S3 (strongest):**
  - Core of the field is roughly **37–38 s**, with some much slower outliers.
  - Car #2’s dot is near the **lower (faster) edge** of the box, indicating it’s **one of the stronger cars in S3**.
  - This sector is a **relative strength** and can be used as a key place to attack and consolidate gains.

- **Overall takeaway:**
  - Car #2 is **competitively paced in all sectors**, with:
    - **S1** = main weakness,
    - **S2** = slightly positive / average+,
    - **S3** = clear strength.
  - Driver coaching and setup work should **prioritize S1**, while **protecting S3 performance**.

```python
def rank_column(df, col):
    # lower = better
    return (
        df[col]
        .rank(method="min")
        .loc[df["is_our_car"]]
        .iloc[0]
    )

def summarize_sector_race(sector_stats, race_label):
    return {
        "car_number": CAR_NUM,
        "race": race_label,
        "S1_mean_s": sector_stats.loc[sector_stats["is_our_car"], "S1_s"].iloc[0],
        "S1_mean_rank": rank_column(sector_stats, "S1_s"),
        "S2_mean_s": sector_stats.loc[sector_stats["is_our_car"], "S2_s"].iloc[0],
        "S2_mean_rank": rank_column(sector_stats, "S2_s"),
        "S3_mean_s": sector_stats.loc[sector_stats["is_our_car"], "S3_s"].iloc[0],
        "S3_mean_rank": rank_column(sector_stats, "S3_s"),
    }

summary_r1 = summarize_sector_race(sector_stats_r1, "R1")
summary_r2 = summarize_sector_race(sector_stats_r2, "R2")

driver_vir = pd.DataFrame([summary_r1, summary_r2])
driver_vir
```

```python
CAR_NUM = 2  # our car

def make_driver_row(sector_stats, race_label: str) -> dict:
    # rank helpers
    def rank_for(col: str) -> float:
        return (
            sector_stats[col]
            .rank(method="min")  # 1 = best
            .loc[sector_stats["car_number"] == CAR_NUM]
            .iloc[0]
        )

    row = sector_stats.loc[sector_stats["car_number"] == CAR_NUM].iloc[0]

    return {
        "car_number": CAR_NUM,
        "race": race_label,
        "S1_mean_s": row["S1_mean_s"],
        "S1_mean_rank": rank_for("S1_mean_s"),
        "S2_mean_s": row["S2_mean_s"],
        "S2_mean_rank": rank_for("S2_mean_s"),
        "S3_mean_s": row["S3_mean_s"],
        "S3_mean_rank": rank_for("S3_mean_s"),
    }

driver_vir = pd.DataFrame([
    make_driver_row(sector_stats_r1, "R1"),
    make_driver_row(sector_stats_r2, "R2"),
])

driver_vir
```

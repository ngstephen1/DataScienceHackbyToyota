```python
import pandas as pd
from pathlib import Path

%matplotlib inline

PROC = Path("../data/processed/barber")

r1 = pd.read_csv(PROC / "barber_r1_car2_sector_summary.csv")
r2 = pd.read_csv(PROC / "barber_r2_car2_sector_summary.csv")

r1["race"] = "R1"
r2["race"] = "R2"

driver_barber = pd.concat([r1, r2], ignore_index=True)
driver_barber
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(1, 3, figsize=(15, 4), sharey=False)

for ax, sec in zip(axes, ["S1", "S2", "S3"]):
    ax.bar(driver_barber["race"], driver_barber[f"{sec}_mean_s"])
    ax.set_title(f"{sec} mean time vs race")
    ax.set_ylabel("time (s)")

plt.suptitle("Barber â€“ Car #2 sector pace R1 vs R2")
plt.tight_layout()
plt.show()
```

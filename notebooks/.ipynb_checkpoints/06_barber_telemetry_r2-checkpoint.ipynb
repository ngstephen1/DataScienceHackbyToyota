```python
from pathlib import Path
import pandas as pd

CAR_ID = "GR86-002-000"
CAR_PATH = Path(f"../data/raw/barber/R1_barber_telemetry_vehicle_{CAR_ID}.csv")

print("Exists? ->", CAR_PATH.exists())

df_car = pd.read_csv(CAR_PATH, low_memory=False)
df_car.head()
```

```python
import numpy as np

# Ensure telemetry_value is numeric
df_car["telemetry_value"] = pd.to_numeric(df_car["telemetry_value"], errors="coerce")

# Pivot: index = (lap, timestamp), columns = telemetry_name
wide = (
    df_car
    .pivot_table(
        index=["lap", "timestamp"],
        columns="telemetry_name",
        values="telemetry_value",
        aggfunc="last",
    )
    .reset_index()
)

wide.columns.name = None  # remove pivot name
wide = wide.sort_values(["lap", "timestamp"]).reset_index(drop=True)

wide.head()
```

```python
wide["timestamp_dt"] = pd.to_datetime(wide["timestamp"])
wide[["lap", "timestamp", "timestamp_dt"]].head()
```

```python
import matplotlib.pyplot as plt

lap_example = int(wide["lap"].min())
lap_df = wide[wide["lap"] == lap_example].copy()

print("Plotting lap", lap_example, "with", len(lap_df), "samples")
```

```python
fig, ax = plt.subplots(figsize=(10, 4))

ax.plot(lap_df["timestamp_dt"], lap_df["aps"], label="aps (throttle)")
ax.plot(lap_df["timestamp_dt"], lap_df["pbrake_r"], label="pbrake_r (rear brake)")

ax.set_title(f"Barber R1  {CAR_ID}  Lap {lap_example}")
ax.set_xlabel("Time")
ax.set_ylabel("Value")
ax.legend()
ax.grid(True)
plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 4))
ax.plot(lap_df["timestamp_dt"], lap_df["Laptrigger_lapdist_dls"])
ax.set_title(f"Barber R1  {CAR_ID}  Lap {lap_example}  Distance from SF")
ax.set_xlabel("Time")
ax.set_ylabel("Distance (m)")
ax.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Basic lap time from telemetry (for this car)
lap_summary = (
    wide.groupby("lap")["timestamp_dt"]
        .agg(["min", "max"])
        .rename(columns={"min": "lap_start_time", "max": "lap_end_time"})
)

lap_summary["lap_time_s"] = (
    lap_summary["lap_end_time"] - lap_summary["lap_start_time"]
).dt.total_seconds()

# Optional extra features if columns exist
for col in ["Speed", "aps", "pbrake_f", "pbrake_r"]:
    if col in wide.columns:
        lap_summary[f"{col}_mean"] = wide.groupby("lap")[col].mean()
        lap_summary[f"{col}_max"]  = wide.groupby("lap")[col].max()

lap_summary.head()
```

```python
fig, ax = plt.subplots(figsize=(10, 4))
ax.plot(lap_summary.index, lap_summary["lap_time_s"], marker="o")
ax.set_title(f"Barber R1  {CAR_ID}  Lap time from telemetry")
ax.set_xlabel("Lap")
ax.set_ylabel("Lap time (s)")
ax.grid(True)
plt.tight_layout()
plt.show()
```

### Interpretation  Lap time vs Lap (Barber R1, GR86-002-000)

- **Lap 2 spike**: One lap is dramatically slower than the rest. This is likely an
  **out-lap, safety-car lap, or major event (traffic/off)** rather than genuine pace.
- **Stable race stint**: From about **lap 3 to lap 28** the lap times are very
  consistent, suggesting:
  - the driver is running at a stable race pace, and  
  - tire degradation over this window is modest (no big upward drift in laptime).
- **Short final lap**: The last lap is much shorter than the rest, which is
  consistent with an **in-lap** (box after finish) or a partial cooldown lap.
- For our strategy engine, this tells us:
  - We should treat laps like **Lap 2 and the final lap as non-representative**
    (exclude from pace modelling).
  - The **true race pace window** for this car at Barber R1 is roughly laps 328.

```python
if "pbrake_f" in wide.columns:
    brake_mean = wide.groupby("lap")["pbrake_f"].mean()

    fig, ax = plt.subplots(figsize=(10, 4))
    ax.plot(brake_mean.index, brake_mean.values, marker="o")
    ax.set_title(f"Barber R1  {CAR_ID}  Avg front brake pressure by lap")
    ax.set_xlabel("Lap")
    ax.set_ylabel("Mean pbrake_f (bar)")
    ax.grid(True)
    plt.tight_layout()
    plt.show()
```

### Interpretation  Avg Front Brake Pressure by Lap (pbrake_f)

- **Early-lap caution**: The first few laps show **lower mean front brake pressure**,
  consistent with the driver bringing tires and brakes up to temperature and
  leaving a bit of margin.
- **Mid-stint ramp-up**: From mid-race onwards, the mean `pbrake_f` rises and then
  oscillates around a higher level. This suggests:
  - growing driver confidence (braking later/harder),
  - lower fuel load allowing deeper braking zones.
- **Lap-to-lap variability**: Individual spikes/dips in mean brake pressure likely
  correspond to:
  - traffic (backing off for slower cars),
  - mistakes or small offs,
  - or local yellow/safety-car segments.
- For the driver-insight layer, this pattern is useful because we can:
  - compare **braking behaviour between stints / races**, and
  - flag laps where braking behaviour deviates significantly from the
    drivers normal profile at Barber.

```python
from pathlib import Path
import pandas as pd

CAR_ID = "GR86-002-000"
CAR_PATH = Path(f"../data/raw/barber/R1_barber_telemetry_vehicle_{CAR_ID}.csv")

print("Exists? ->", CAR_PATH.exists())
df_car = pd.read_csv(CAR_PATH, low_memory=False)
df_car.head()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# make telemetry_value numeric
df_car["telemetry_value"] = pd.to_numeric(df_car["telemetry_value"], errors="coerce")

wide = (
    df_car
    .pivot_table(
        index=["lap", "timestamp"],
        columns="telemetry_name",
        values="telemetry_value",
        aggfunc="last",
    )
    .reset_index()
)

wide.columns.name = None
wide = wide.sort_values(["lap", "timestamp"]).reset_index(drop=True)

wide["timestamp_dt"] = pd.to_datetime(wide["timestamp"], errors="coerce")

wide.head()
```

```python
[c for c in wide.columns if "Laptrigger" in c or "lapdist" in c.lower()]
```

```python
lap_len = barber_meta.circuit_length_m
s1, s2, s3 = barber_meta.sector_lengths_m
s1_end = s1
s2_end = s1 + s2
s3_end = s1 + s2 + s3

if "Laptrigger_lapdist_dls" in wide.columns:
    dist = wide["Laptrigger_lapdist_dls"]

    conditions = [
        dist <= s1_end,
        (dist > s1_end) & (dist <= s2_end),
        dist > s2_end,
    ]
    choices = ["S1", "S2", "S3"]

    wide["sector"] = np.select(conditions, choices, default="UNKNOWN")
    wide["sector"].value_counts()
else:
    print("No Laptrigger_lapdist_dls column found  cannot do sector timing here.")
```

```python
# sector start/end and duration per (lap, sector)
sector_summary = (
    wide.dropna(subset=["sector"])
        .groupby(["lap", "sector"])["timestamp_dt"]
        .agg(["min", "max"])
        .rename(columns={"min": "sec_start", "max": "sec_end"})
)

sector_summary["sector_time_s"] = (
    sector_summary["sec_end"] - sector_summary["sec_start"]
).dt.total_seconds()

sector_summary.head()
```

```python
sector_pivot = sector_summary["sector_time_s"].unstack("sector")
sector_pivot.head()
```

```python
fig, ax = plt.subplots(figsize=(10, 4))

for sec in ["S1", "S2", "S3"]:
    if sec in sector_pivot.columns:
        ax.plot(sector_pivot.index, sector_pivot[sec],
                marker="o", label=sec)

ax.set_title(f"Barber R1  {CAR_ID}  Sector times by lap")
ax.set_xlabel("Lap")
ax.set_ylabel("Sector time (s)")
ax.grid(True)
ax.legend()
plt.tight_layout()
plt.show()
```

### Interpretation  Sector Times by Lap (Barber R1, GR86-002-000)

- **Sector identities:**
  - **S1** (blue) is the shortest sector and sits around the low-to-mid 2030s range
    on normal laps.
  - **S2** (orange) is consistently longer ( 40 s), which matches the geometry:
    mid-lap flowing section with several corners.
  - **S3** (green) should be similar in length to S1/S2 for clean race laps, but we
    see several laps where S3 time jumps above **100 s**.

- **Abnormal S3 spikes & UNKNOWN column:**
  - Those huge S3 times and the `UNKNOWN` column in the table are a sign that some
    laps include **pit-lane running, safety-car segments, or partial laps**, which
    make one sector absorb a very long time.
  - These laps are **not representative of pure pace** and should be excluded when
    building driver-pace or tire-degradation models.

- **Usable pace window:**
  - For S1 and S2, once we ignore the early odd laps, the sector times stabilise
    nicely, giving a clean **race-pace window** we can use in the driver-insight and
    strategy models.
  - In later stages, well filter laps where all sectors fall within a reasonable
    band (e.g., S1 < 60 s, S2 < 80 s, S3 < 80 s) and treat the rest as **event
    laps (pit / yellow / errors)**.

```python
# mark laps where all sectors are within a "normal" band
sector_clean = sector_pivot.copy()

# you can tweak these thresholds later
good_mask = (
    (sector_clean["S1"] < 60) &
    (sector_clean["S2"] < 80) &
    (sector_clean["S3"] < 80)
)

sector_clean["is_race_pace_lap"] = good_mask
sector_clean.head()
```

```python
import pandas as pd

# lap start/end and lap_time_s from telemetry
lap_summary = (
    wide.groupby("lap")["timestamp_dt"]
        .agg(["min", "max"])
        .rename(columns={"min": "lap_start_time", "max": "lap_end_time"})
)

lap_summary["lap_time_s"] = (
    lap_summary["lap_end_time"] - lap_summary["lap_start_time"]
).dt.total_seconds()

# (optional) add simple pace features if columns exist
for col in ["Speed", "aps", "pbrake_f", "pbrake_r"]:
    if col in wide.columns:
        lap_summary[f"{col}_mean"] = wide.groupby("lap")[col].mean()
        lap_summary[f"{col}_max"]  = wide.groupby("lap")[col].max()

lap_summary.head()
```

```python
if "pbrake_f" in wide.columns:
    brake_mean = wide.groupby("lap")["pbrake_f"].mean()
else:
    brake_mean = None

brake_mean.head() if brake_mean is not None else "no pbrake_f column"
```

```python
from pathlib import Path

# lap_summary came from earlier (lap_time_s, maybe aps_mean / pbrake_f_mean, etc.)
# brake_mean we computed as mean pbrake_f by lap
# sector_clean has S1/S2/S3 sector times and is_race_pace_lap

lap_features = lap_summary.copy()

# attach brake_mean if it exists
if "brake_mean" in globals() and brake_mean is not None:
    lap_features["pbrake_f_mean"] = brake_mean

# attach sector times + race-pace flag
for col in ["S1", "S2", "S3", "is_race_pace_lap"]:
    if col in sector_clean.columns:
        new_name = {
            "S1": "sector1_time_s",
            "S2": "sector2_time_s",
            "S3": "sector3_time_s",
        }.get(col, col)
        lap_features[new_name] = sector_clean[col]

lap_features.head()
```

```python
OUT_DIR = Path("../data/processed/barber")
OUT_DIR.mkdir(parents=True, exist_ok=True)

out_path = OUT_DIR / f"barber_r1_{CAR_ID}_lap_features.csv"
lap_features.to_csv(out_path, index_label="lap")

out_path
```
